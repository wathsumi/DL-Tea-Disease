```python

import os
import numpy as np
import math
import shutil

def dataFolder(p, split):
    # Create a "train" folder if it doesn't exist
    if not os.path.exists("./"+p):
        os.mkdir("./"+p)

    # Define the root directory of your dataset
    ROOT_DIR = "/content/tea sickness dataset"

    # Dictionary to store the number of images per class
    number_of_images = {}

    # Iterate through subdirectories (classes)
    for class_dir in os.listdir(ROOT_DIR):
        class_path = os.path.join(ROOT_DIR, class_dir)

        # Check if the path is a directory
        if os.path.isdir(class_path):
            # Count the number of images in the class directory
            image_count = len(os.listdir(class_path))
            number_of_images[class_dir] = image_count

            # Create a subdirectory in the "train" folder for this class
            train_class_dir = os.path.join("./"+p+"/", class_dir)
            os.makedirs(train_class_dir, exist_ok=True)

            # Randomly copy a subset of images to the "train" folder
            num_images_to_copy = math.floor(0.7 * image_count) - 8
            if num_images_to_copy > 0:
                class_images = os.listdir(class_path)
                selected_images = np.random.choice(class_images, size=num_images_to_copy, replace=False)
                for img in selected_images:
                    source_path = os.path.join(class_path, img)
                    destination_path = os.path.join(train_class_dir, img)
                    shutil.copy(source_path, destination_path)

    # Print the number of images per class
    for class_name, count in number_of_images.items():
        print(f"Class: {class_name}, Number of Images: {count}")

    print(f"{p} Folder exists")

# Example usage:
dataFolder("train", 0.7)
```

preparing our data using Data Generator

```python
# Define your train, validation, and test directories
train_dir = "/content/train"
val_dir = "/content/val"
test_dir = "/content/test"
```

```python
# ...

# Data preprocessing and augmentation
train_data_gen = ImageDataGenerator(
    rescale=1.0 / 255.0,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)

val_test_data_gen = ImageDataGenerator(rescale=1.0 / 255.0)

train_data = train_data_gen.flow_from_directory(
    train_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'  # Change this to 'categorical' if you have multiple classes
)

val_data = val_test_data_gen.flow_from_directory(
    val_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'  # Change this to 'categorical' if you have multiple classes
)

# ...

# Model architecture
model = Sequential()

model.add(Conv2D(filters=16, kernel_size=(3, 3), activation='relu', input_shape=(224, 224, 3)))
# Add more layers as needed...

model.add(Flatten())  # Add this layer before the output layer for flattening the convolutional features

model.add(Dense(1, activation='sigmoid'))  # Output layer for binary classification

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# ...

# Early stopping and model checkpoint
es = EarlyStopping(monitor="val_accuracy", min_delta=0.01, patience=3, verbose=1, mode='auto')
mc = ModelCheckpoint(monitor="val_accuracy", filepath="./bestmodel.h5", verbose=1, save_weights_only=True, mode='auto')

cd = [es, mc]  # Define the list of callbacks

# Model training
hs = model.fit(
    train_data,
    steps_per_epoch=len(train_data),
    epochs=30,
    verbose=1,
    validation_data=val_data,
    validation_steps=len(val_data),
    callbacks=cd  # Pass the callbacks to the fit method
)

```

```python
# Model Graphical Interpretatin

h = hs.history
keys = h.keys()
print(keys)
```

```python
import matplotlib.pyplot as plt

plt.plot(h['val_accuracy'])
plt.plot(h['val_accuracy'], c = "red")

plt.title("acc vs val-acc")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot(h['loss'])
plt.plot(h['val_loss'], c = "red")

plt.title("loss vs val-loss")
plt.show()
```
